```python
#1

import csv
def LerInformacao (filename):
    header=[]
    fichaPacientes = []

    with open(filename , 'r', encoding="utf-8") as file:
        csvreader = csv.reader(file)
        header = next(csvreader)
        
        for paciente in csvreader:
            fichaPacientes.append(paciente)
              
    return fichaPacientes
```

```python
LerInformacao('myheart.csv')
```

```python
# número de doentes

import csv
def doenca(filename):
    fichaPacientes=LerInformacao(filename)
    doente=0
    
    for paciente in fichaPacientes:
        idade , sexo , tensão , colesterol , batimento , temDoença = paciente
        if temDoença == '1':
            doente = doente + 1

    return doente
```

```python
doenca('myheart.csv')
```

```python
#2
#TABELA
import csv
def sexo(filename):
    fichaPacientes=LerInformacao(filename)
    
    resM = 0
    resF = 0
    
    for paciente in fichaPacientes:
        
        idade , sexo , tensão , colesterol , batimento , temDoença = paciente
        
        if sexo == 'M' and temDoença=='1':
            resM= resM + 1
            
        elif sexo == 'F' and temDoença == '1':
             resF= resF + 1
    x=' '         
    print(5*x,'FEMININO', 5*x, 'MASCULINO')        
    print (8*x , resF,10*x ,resM)         
    return 
        
```

```python
sexo('myheart.csv')
```

```python
# criar dicionário de toda a informação do ficheiro

def dicionario(filename):
    dict={}
    id=1
    fichaPacientes = LerInformacao(filename)
    for paciente in fichaPacientes:
        dict.update({id:paciente})
        id=id+1
        
    return dict
```

```python
dicionario('myheart.csv')
```

```python
#3

import csv
def idade(filename):
    idPacientes=dicionario(filename)
    listaDoentes= []
    
    for i in idPacientes.values():
        if i[5]=='1':
            listaDoentes.append(i) 
            
    divisao_idade = {}
    
    for doente in listaDoentes:
        idade , sexo , tensão , colesterol , batimento , temDoença = doente
        intervalo = int(int(idade)/5)
        
        if divisao_idade.get(intervalo)== None:
            divisao_idade[intervalo]= [doente]
        else:
            divisao_idade[intervalo].append(doente)

    for k,v in sorted(divisao_idade.items()):
        print(k*5 ,'-', k*5 + 4, '»' , len(v))
        
    return 
```
